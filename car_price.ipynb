```python
# manually encoding one column
# Petrol replaced with 0, Diesel with 1, CNG with 2
car_data.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)
# one hot encoding two columns, creates new dummy columns and drops original column
car_data = pd.get_dummies(car_data, columns=['Seller_Type', 'Transmission'], drop_first=True)
```

```python
# data has changed and got new columns
car_data.head()
```

```python
# heat map to find the correlation between the different cols
plt.figure(figsize=(10,7))
sns.heatmap(car_data.corr(), annot=True)
plt.title('Correlation Between Columns')
plt.show()
```

```python
# to see the correlation better, we are plotting another graph
fig=plt.figure(figsize=(7,5))
plt.title('Correlation Between Present Price and Selling Price')
sns.regplot(x='Present_Price', y='Selling_Price', data=car_data)
```

```python
X = car_data.drop(['Car_Name', 'Selling_Price'], axis=1)
y = car_data['Selling_Price']
```

```python
print("Shape of X is: ",X.shape)
print("Shape of y is: ", y.shape)
```

```python
# 30% is test data, 70% is training data
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
print("X_test shape:", X_test.shape)
print("X_train shape:", X_train.shape)
print("y_test shape: ", y_test.shape)
print("y_train shape:", y_train.shape)
```

```python
# normalize the data with StandardScalar
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# load LinearRegression models and fit the training sets to the model
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
# create a variable to store the predicted values
pred = model.predict(X_test)
```

```python
# to see how well the value fits the data, we calc the MSE, MAE, R2
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# print the results
# between the predicted values and the y-test
# MAE, MSE used to determine how close calculated values are to the actual values
# R2 is an indicator of how well the models fits the data, the higher, the better.
print("MAE: ", (metrics.mean_absolute_error(pred, y_test)))
print("MSE: ", (metrics.mean_squared_error(pred, y_test)))
print("R2 score: ", (metrics.r2_score(pred, y_test)))
```

```python
# regression plot between the predicted and actual values
sns.regplot(x=pred, y=y_test)
plt.xlabel("Predicted Price")
plt.ylabel('Actual Price')
plt.title("Actual vs Predicted Price")
plt.show()
```
